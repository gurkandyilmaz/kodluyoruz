```python
csv_2005_2007 = '/kaggle/input/2000-16-traffic-flow-england-scotland-wales/accidents_2005_to_2007.csv'
csv_2009_2011 = '/kaggle/input/2000-16-traffic-flow-england-scotland-wales/accidents_2009_to_2011.csv'
csv_2012_2014 = '/kaggle/input/2000-16-traffic-flow-england-scotland-wales/accidents_2012_to_2014.csv'
```

```python
%%time
accidents_2007 = pd.read_csv(csv_2005_2007, low_memory=False).drop(columns=['Junction_Detail','Junction_Control'])
accidents_2011 = pd.read_csv(csv_2009_2011, low_memory=False).drop(columns=['Junction_Detail','Junction_Control'])
accidents_2014 = pd.read_csv(csv_2012_2014, low_memory=False).drop(columns=['Junction_Detail','Junction_Control'])
```

```python
%%time
accidents_2007['datetime'] = pd.to_datetime(accidents_2007['Date'] +' '+ accidents_2007['Time'])
accidents_2011['datetime'] = pd.to_datetime(accidents_2011['Date'] +' '+ accidents_2011['Time'])
accidents_2014['datetime'] = pd.to_datetime(accidents_2014['Date'] +' '+ accidents_2014['Time'])
```

```python
figure, axes = plt.subplots(3,1, figsize=(24,12))
axes[0].hist(accidents_2007['Day_of_Week'], bins=7, alpha=0.5, rwidth=0.55)
axes[0].set_title('Distribution of Accident numbers with respect to Day of the week')
axes[0].set_xlabel('Day of the Week')
axes[0].set_ylabel('Number of Accidents')

axes[1].hist(accidents_2011['Day_of_Week'], bins=7, alpha=0.5, rwidth=0.55)
axes[1].set_xlabel('Day of the Week')
axes[1].set_ylabel('Number of Accidents')

axes[2].hist(accidents_2014['Day_of_Week'], bins=7, alpha=0.5, rwidth=0.55)
axes[2].set_xlabel('Day of the Week')
axes[2].set_ylabel('Number of Accidents')

plt.show()
```

- There is not much difference in terms of the number of accidents by year.
- In terms of the day of the week, the most accidents occured on Saturdays
- The least amount of accidents occured on the first day of the week, Monday

```python
figure, axes = plt.subplots(3,1, figsize=(24,12))
axes[0].hist(accidents_2007['datetime'].dt.day, bins=30, alpha=0.5, rwidth=0.55, label='2007')
axes[0].set_title('Distribution of Accident numbers with respect to Day of the month')
axes[0].set_xlabel('Day of the Week')
axes[0].set_ylabel('Number of Accidents')
axes[0].legend()

axes[1].hist(accidents_2011['datetime'].dt.day, bins=30, alpha=0.5, rwidth=0.55, label='2011')
axes[1].set_xlabel('Day of the Week')
axes[1].set_ylabel('Number of Accidents')
axes[1].legend()

axes[2].hist(accidents_2014['datetime'].dt.day, bins=30, alpha=0.5, rwidth=0.55, label='2014')
axes[2].set_xlabel('Day of the Week')
axes[2].set_ylabel('Number of Accidents')
axes[2].legend()

plt.show()
```

- Again there is not much difference in terms of the accident counts by year.
- Most accidents occured on the last days of the month. It could be the effect of summing all the accidents from the days 30 and 31.

```python
figure, axes = plt.subplots(3,1, figsize=(24,12))
axes[0].hist(accidents_2007['datetime'].dt.hour, bins=24, alpha=0.5, rwidth=0.55, label='2007')
axes[0].set_title('Distribution of Accident numbers with respect to hour of the day')
axes[0].set_xlabel('Day of the Week')
axes[0].set_ylabel('Number of Accidents')
axes[0].legend()

axes[1].hist(accidents_2011['datetime'].dt.hour, bins=24, alpha=0.5, rwidth=0.55, label='2011')
axes[1].set_xlabel('Day of the Week')
axes[1].set_ylabel('Number of Accidents')
axes[1].legend()

axes[2].hist(accidents_2014['datetime'].dt.hour, bins=24, alpha=0.5, rwidth=0.55, label='2014')
axes[2].set_xlabel('Day of the Week')
axes[2].set_ylabel('Number of Accidents')
axes[2].legend()

plt.show()
```

- The most accidents occur in between 15 pm and 18 pm. People usually are tired and on the way to their home at those times.
- As expected there is a relatively small amount of accidents in the early hours of the day.
- Again we see the similar pattern from year to year

```python
figure, axes = plt.subplots(1,3, figsize=(24,6))
axes[0].bar(x=accidents_2007['Urban_or_Rural_Area'].unique(), height=accidents_2007['Urban_or_Rural_Area'].value_counts(), label='2007')
axes[0].set_xticks([1.0, 2.0])
axes[0].set_xticklabels(['Urban', 'Rural'])
axes[0].set_ylabel('Number of Accidents')
axes[0].legend()
axes[0].set_title('Number of Accidents with respect to crash site')

axes[1].bar(x=accidents_2011['Urban_or_Rural_Area'].unique(), height=accidents_2011['Urban_or_Rural_Area'].value_counts(), label='2011')
axes[1].set_xticks([1.0, 2.0])
axes[1].set_xticklabels(['Urban', 'Rural'])
axes[1].set_ylabel('Number of Accidents')
axes[1].legend()
axes[1].set_title('Number of Accidents with respect to crash site')

axes[2].bar(x=accidents_2014['Urban_or_Rural_Area'].unique(), height=accidents_2014['Urban_or_Rural_Area'].value_counts(), label='2014')
axes[2].set_xticks([1.0, 2.0])
axes[2].set_xticklabels(['Urban', 'Rural'])
axes[2].set_ylabel('Number of Accidents')
axes[2].legend()
axes[2].set_title('Number of Accidents with respect to crash site')

plt.show()
```

- The most accidents occured in Urban areas where the population density is larger than that of Rural areas.
- As the year passed, the number of accidents in Rural areas decreased from ~200k to ~150k. This could be due to the fact that people were leaving their villages to find a better opportunities in Urban regions.
